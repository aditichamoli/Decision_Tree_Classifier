<h1>Prediction using Decision Tree Algorithm</h1>

<h1>Importing libraries</h1>

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import sklearn.tree as tree
```

<h1>Importing dataset</h1>

```python
iris_data=pd.read_csv("Iris.csv")
iris_data
```

<h1>Splitting the data into testing and training data for model building</h1>

```python
X=iris_data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y=iris_data['Species']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

<h1>Fitting the data into DecisionTree Classifier model</h1>

```python
decision_tree=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
decision_tree.fit(x_train,y_train)
```

<h1>Visualizing the Decision Tree</h1>

```python
dot_data = tree.export_graphviz(decision_tree, out_file=None, 
                         feature_names=X.columns,  
                         class_names=["Iris-virginica","Iris-setosa","Iris-Versicolor"],  
                         filled=True, rounded=True,  
                         special_characters=True,proportion=True)
```

```python
import pydotplus
graph = pydotplus.graph_from_dot_data(dot_data)
from IPython.display import Image
Image(graph.create_png())
```

<h1>Performing predictions</h1>

```python
y_prediction = decision_tree.predict(x_test)
```

<h1>Comparing actual values vs predicted values</h1>

```python
df= pd.DataFrame({'Actual':y_test,'Predicted':y_prediction})
df
```

<h1> Measuring the accuracy of the DecisionTree Classifier </h1>

```python
accuracy_score(y_true=y_test,y_pred=y_prediction)
```
